```python
# Define the confidence interval function.

def confInterval(x, confInt, meanOrmedian):
    if meanOrmedian == 0:
        center = np.median(x)
    elif meanOrmedian == 1:
        center = np.mean(x)
        
    lowerLim = np.percentile(x, 100*(1-confInt)/2)
    upperLim = np.percentile(x, 100*(1+confInt)/2)
    
    return(center, upperLim-center, center-lowerLim)        
```

3. Poisson Confidence Intervals

3a)

According to Gehrels (1985), the two-sided upper ($\lambda_u$) and lower ($\lambda_l$) limits for a Poisson-like distribution are defined by, respectively:

$$\sum_{x = 0}^{n} \frac{\lambda_u^x e^{-\lambda_u}}{x!} = 1 - \frac{(1 + CL^{\prime})}{2},$$

$$\sum_{x = 0}^{n - 1} \frac{\lambda_l^x e^{\lambda_l}}{x!} = \frac{(1 + CL^{\prime})}{2},$$

where $CL^{\prime}$ is the "confidence interval" (e.g. 99%).

We use the information provided in class:

$$\sum_{j = 0}^{k - 1} \frac{\lambda^j e^{-\lambda}}{j!} = 1 - \frac{\int_{0}^{\lambda} e^{-t} t^{k-1} dt}{\Gamma(k)},$$

to determine the upper and lower limits. For the case of the upper limit, we define $n = k - 1$, use the equation provided in class, just above of this paragraph, and in the final expression, we substitute $k = n + 1$:

$$ CL^{\prime} = 2 \times \frac{\int_{0}^{\lambda_u} e^{-t} t^{k - 1} dt}{\Gamma(k)} - 1 (*)$$

$$ CL^{\prime} = 2 \times (1 - \frac{\int_{0}^{\lambda_l} e^{-t} t^{k - 1} dt}{\Gamma(k)}) - 1 (**)$$

```python
# Define the gamma function.

def Gamma(t, n):
    G = np.exp(-t) * t**(n-1)
    return G

def GammaMod(t, n):
    G = np.exp(-t) * t**(n)
    return G
```

In order to find the upper and lower limits, I will create a lot of values of both numbers, plug them into Equations $(*)$ and $(**)$, and require them to return the wanted confidence limit $CL^{\prime}$, which is arbitrary and we choose it by hand. (There is probably a better way of doing this, perhaps by taking some sort of transform, but I cannot find, or think of, anyway of doing that.)

```python
def ComparisonFunctionUpper(n, lambd, CLprime):
    # Upper limit finder:
    lambd_dummy = 0;
    for i in lambd:
        CLprime_test =  (2*(quad(GammaMod, 0, i, n)[0])/quad(GammaMod, 0, np.inf, n)[0]) - 1
        #print(CLprime_test)
        if abs(CLprime_test - CLprime) < 0.001:
            lambd_dummy = i
            break
    return lambd_dummy

def ComparisonFunctionLower(n, lambd, CLprime):
    # Lower limit finder:
    lambd_dummy = 0;
    for i in lambd:
        CLprime_test = (2 - 2*(quad(Gamma, 0, i, n)[0])/quad(Gamma, 0, np.inf, n)[0]) - 1
        #print(CLprime_test)
        if abs(CLprime_test - CLprime) < 0.001:
            lambd_dummy = i
            break
    return lambd_dummy
```

According to Gehrels (1985), the two-sided upper ($p_u$) and lower ($p_l$) limits for a Binomial-like distribution are defined by, respectively:

where $CL^{\prime}$ is the "confidence interval" (e.g. 99%).

5. Analytic Least-Squares Fitting

5a) For $g(x) = mx + c$, derive the formulas for $c$, $m$, $\sigma_c$, and $\sigma_m$

I will base the following derivation in the procedure shown in in John R. Taylor's "An introduction to error analysis: the study of uncertainties in physical measurements" (1997), the Wolfram webpage on "Least squares fitting": https://mathworld.wolfram.com/LeastSquaresFitting.html, and the lecture notes "Lecture 10-12A Least Squares Basics" and "Lecture 10-12B Linear Least Squares". I will explain each step as much as I can

(i) Find $c$ and $m$

As explained in the lecture notes, the general approach to finding the values $c$ and $m$ from $g(x) = mx + c$, is by minimizing the $\chi^2 = \sum_{i = 1}^{N} \left(\frac{y_i - f(x_i; a_j)}{\sigma_i}\right)^2$ with respect to the variables $a_j$. In this case, the correct expression for the $\chi^2$ value is:

$$\chi^2 = \sum_{i = 1}^{N} w_i \left(y_i - c - mx_i\right)^2,$$

where $w_i$ are defined as the "weights" in the "$y_i$" variable, wich is just $g_i = g(x_i)$: $w_i \equiv \frac{1}{\sigma_i^2}$.

We assume that the uncertainties in the $x$ variable are negligible, or not dominant. We can always manipulate the equation in such a way that the dominant uncertainties $\sigma_i$ correspond to the "vertical" variable.

Now we minimize:

$$\frac{\partial \chi^2}{\partial c} = - 2 \sum_{i = 1}^{N} w_i \left(y_i - c - mx_i \right) = 0,$$

$$\frac{\partial \chi^2}{\partial m} = - 2 \sum_{i = 1}^{N} w_i x_i \left(y_i - c - mx_i \right) = 0,$$

which can be written as:

$$ \sum{w_i y_i} = c \sum{w_i} + m \sum{w_i x_i}$$

$$ \sum{w_i x_i y_i} = c \sum{w_i x_i} + m \sum{w_i x_i^2}$$

We solve solve for $c$ and equate to find $m$. Then we substitute this value of $m$ in either equation solved for $c$ and we find $c$. (I did this by hand, but I include here a summary of the procedure):

$$c = \frac{\sum{wy} - m \sum{wx}}{\sum{w}}$$

$$c = \frac{\sum{wxy} - m \sum{wx^2}}{\sum{wx}}$$

Equating these two expressions and solving for $m$ gives:

$$m = \frac{\sum{w} \sum{wxy} - \sum{wy} \sum{wx}}{\Delta} \; (\textrm{v}),$$

where $\Delta \equiv \sum{w} \sum{wx^2} - \left(\sum{wx}\right)^2$.

we plug this back into either expression for $c$:

$$c = \frac{\sum{wx^2} \sum{wy} - \sum{wxy} \sum{wx}}{\Delta} \; (\textrm{vi})$$.

As an answer to the first part of this problem, the values for $c$ and $m$ are given by Euqations (vi) and (v), respectively.

(ii) Find the uncertainties in $c$ and $m$: $\sigma_c$ and $\sigma_m$, respectively

The uncertainties in $c$ and $m$ are given by simple error propagation.
